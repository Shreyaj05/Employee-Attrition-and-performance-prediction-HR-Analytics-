# Employee Attrition and performance prediction (HR Analytics)

## Description:

### This project aims to predict employee attrition in an organization using supervised machine learning techniques. Attrition, or employee turnover, is a critical issue in Human Resource (HR) analytics, and early prediction helps in strategic decision-making for employee retention.  A real-world HR dataset was used, containing various employee attributes such as age, gender, job satisfaction, monthly income, distance from home, and more. The target variable is Attrition, classified as Yes or No.

## Feature Engineering 

```python
data.drop(columns=['EmployeeCount','Over18','StandardHours','EmployeeNumber'] , axis=1 ,inplace=True)
```

```python
x=data.drop(columns=['Attrition'])
y=data['Attrition']
```

```python
categoricals = x.select_dtypes(include='object').columns
x = pd.get_dummies(x, columns=categoricals, drop_first=True)
```

```python
scaler=StandardScaler()
x=scaler.fit_transform(x)
```

```python
LE=LabelEncoder()
y=LE.fit_transform(y)
```

## Train Test Split

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2)
```

```python
y_train.shape
```

```python
x_train.shape
```

## Model Training and Evaluation


## Logistic Regression

```python
lr = LogisticRegression(max_iter=500, class_weight='balanced')
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
from sklearn.model_selection import cross_val_score
np.mean(cross_val_score(LogisticRegression(),x, y,cv=10,scoring='accuracy'))
```

### Random Forest

```python
rf = RandomForestClassifier(class_weight='balanced')
rf.fit(x_train, y_train)
y_pred = rf.predict(x_test)
from sklearn.model_selection import cross_val_score
np.mean(cross_val_score(RandomForestClassifier(),x, y,cv=10,scoring='accuracy'))
```

### XGBoost

```python
# XGBoost
xgb = XGBClassifier( eval_metric='logloss')
xgb.fit(x_train, y_train)
y_pred = xgb.predict(x_test)
from sklearn.model_selection import cross_val_score
np.mean(cross_val_score(XGBClassifier(eval_metric='logloss'),x, y,cv=10,scoring='accuracy'))
```

### SVC

```python
svm = SVC(probability=True, class_weight='balanced')
svm.fit(x_train, y_train)
y_pred = svm.predict(x_test)
from sklearn.model_selection import cross_val_score
np.mean(cross_val_score(SVC(),x, y,cv=10,scoring='accuracy'))
```

### KNeighbors

```python
# KNN
scores=[]
for i in range(1,16):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    y_pred = knn.predict(x_test)
    scores.append(accuracy_score(y_test,y_pred))
```

```python
plt.plot(range(1,16),scores)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
accuracy_score(y_test,y_pred)
```

### Stacking

```python
estimators = [
    ('rf', RandomForestClassifier(n_estimators=10, random_state =42)),
    ('knn', KNeighborsClassifier(n_neighbors=10)),
    ('gbdt',GradientBoostingClassifier())
]   
```

```python
from sklearn.ensemble import StackingClassifier
clf= StackingClassifier(
    estimators=estimators,
    final_estimator= LogisticRegression(),
    cv=10
)  
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
accuracy_score(y_test,y_pred)
```

### From Above we can easily see that logistic regression is the best fit Algorithm for this dataset as it gives higher accuracy as compared to other algorithms

### Can we reduce this dataset into a few dimensions ?

```python
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
for i in range(1,31):
    pca=PCA(n_components=i)
    x_train_trf=pca.fit_transform(x_train)
    x_test_trf=pca.transform(x_test)
    knn= KNeighborsClassifier()
    knn.fit(x_train_trf,y_train)
    y_pred=knn.predict(x_test_trf)
    print(i,accuracy_score(y_test,y_pred))
```

### As we can see from above that we can reduce this dataset into 2d or 3d dimensions without affecting much on its accuracy
